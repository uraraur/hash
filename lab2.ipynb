```python
n = 32
pad = 128 - n 
```

```python
def generate_bytes(len):
    return random.randbytes(len)


def generate_table(K, L, R):
    X = {}

    for i in range(K):
        xi0 = generate_bytes(n // 8)
        xij = xi0
        for j in range(L):      
            xij = hl.sha256(R(xij)).digest()
        X[xij.hex()] = xi0.hex()

    return X

def preimage_search(K, L, R, X, hash_value):
    y = hash_value
    
    for j in range(L):
        if y.hex() in X:
            print('YUPIE')
            x = X[y.hex()]
            x = bytes.fromhex(x)
            for m in range(L - j):
                x = hl.sha256(R(x)).digest()
            return R(x)
        y = hl.sha256(R(y)).digest()

    return 0
```

```python
N = 10000
K = 2**14
L = 2**7

r = generate_bytes(pad // 8)
def R(x):
    return (x + r)

X = generate_table(K, L, R)
print(X)
for i in range(N):
    x = generate_bytes(256 // 8)
    h_x = hl.sha256(x).digest()
    preim = preimage_search(K, L, R, X, h_x)
    if preim != 0:
        print(f"i: {i}, x: {x.hex()}, h_x: {h_x}, preim: {preim}")
```
